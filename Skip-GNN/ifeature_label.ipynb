```python
## Load Path
get_path = 'drive/MyDrive/Skip-GNN/'
data_path = get_path + 'data/IAV/'
```

```python
import pandas as pd
import numpy as np
```

```python
protein_list = pd.read_csv(data_path + 'protein_list.csv')
# protein_list.columns = ['idx', 'Protein1_ID']
protein_list
```

```python
## hold-out test set
test_df = pd.read_csv(data_path + 'edges/nn_test_idx.csv')

## Training and Validation Set
pos_org = pd.read_csv(data_path + 'edges/nn_pos_idx.csv')
neg_org = pd.read_csv(data_path + 'edges/nn_neg_idx.csv')
```

```python
p_idx_arr = []
```

```python
p_idx_arr.extend(test_df['Protein1_ID'].to_numpy())
p_idx_arr.extend(test_df['Protein2_ID'].to_numpy())

p_idx_arr.extend(pos_org['Protein1_ID'].to_numpy())
p_idx_arr.extend(pos_org['Protein2_ID'].to_numpy())

p_idx_arr.extend(neg_org['Protein1_ID'].to_numpy())
p_idx_arr.extend(neg_org['Protein2_ID'].to_numpy())
```

```python
p_idx_arr_u = list(set(p_idx_arr))
print(len(p_idx_arr_u))
```

```python
new_protein_list = protein_list
```

```python
new_protein_list['ifeat_label'] = 0
```

```python
for i in new_protein_list.index:

  # print(new_protein_list['Unnamed: 0'][i])
  
  if new_protein_list['Unnamed: 0'][i] in p_idx_arr_u:
    new_protein_list['ifeat_label'][i] = 1
  
  # else:
  #   new_protein_list['ifeat_label'][i] = 0
```

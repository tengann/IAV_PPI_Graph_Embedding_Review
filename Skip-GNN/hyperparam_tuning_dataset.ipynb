```python
## UniProt ID - SkipGNN

# neg_proportion = '100p (all)', '50p', '25p', '12-5p (12.5p)' 
# neg_sample_size = 206010, 147150, 58860, 17660, 5886

## Train_Val : Test = 9 : 1
def create_fold(neg_prop, neg_sample_size):

  ## Retrieve raw datasets
  pos_test = pd.read_csv(edges_path + 'unbalanced_dataset/positive_edges_test.csv')

  neg_org = pd.read_csv(edges_path + 'unbalanced_dataset/negative_edges_' + neg_prop + '.csv') ## negative_edges
  neg = shuffle(neg_org)

  ## Sample negative dataset
  neg = neg[:neg_sample_size].reset_index(drop=True)

  neg_test = neg.sample(frac = 0.1, replace = False) 
  neg_train_val = neg[~neg.index.isin(neg_test.index)]
 
  test_df = pd.concat([pos_test, neg_test], ignore_index=True)

  return neg_train_val, test_df
```

```python
neg_train_val, test = create_fold('all', 5886)

print(neg_train_val)
print(test)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

xpos = [1, 2, 3, 4, 5, 6]
ypos = [1, 2, 3, 4, 5, 6]
zpos = np.zeros(6)

dx = np.ones(6)
dy = np.ones(6)
dz = [0.9167, 0.9137, 0.9118, 0.9168, 0.9237, 0.92]

ax1.bar3d(xpos, ypos, zpos, dx, dy, dz)

ax1.set_xlabel('num walks')
ax1.set_ylabel('walk length')
ax1.set_zlabel('F1-Score')

plt.setp(ax1, xticks=[2, 3, 4, 5, 6, 7], xticklabels=['8', '16', '32', '64', '128', '256'])
plt.setp(ax1, yticks=[2, 3, 4, 5, 6, 7], yticklabels=['8', '16', '32', '64', '128', '256'])

locs, labels = plt.xticks()
print(locs, labels)

plt.show()
```

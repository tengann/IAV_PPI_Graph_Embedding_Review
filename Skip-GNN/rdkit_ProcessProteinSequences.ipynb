### Process Protein Sequences

```python
import pandas as pd
```

```python
## Read dataset
df = pd.read_csv(get_path + 'data/all_sequences.txt', sep='\t', header=None)
df.columns=['id', 'seq']
df = df.rename(columns={'id': 'Protein1_ID'})
df
```

```python
## Count sequence length
new_pl = pd.read_csv('/drive/MyDrive/Structure_PPI_GNN/protein_list_filtered.csv')
new_pl
```

```python
protein_list = pd.read_csv('drive/MyDrive/Skip-GNN/data/IAV/protein_list.csv')
protein_list = protein_list.drop(columns=['Unnamed: 0'])
protein_list ## 15685 rows
```

```python
## Exclude sequences not in protein list
merge_df = pd.merge(df, protein_list, on=['Protein1_ID'])
merge_df
```

```python
merge_df.to_csv('drive/MyDrive/Skip-GNN/data/new_all_sequences.txt', index=False, sep='\t', header=None)
```
